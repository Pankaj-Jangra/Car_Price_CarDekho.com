```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=32)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {'n_estimators': [int(x) for x in np.linspace(start=100,stop=1000,num=10)],
               'max_features':["auto","sqrt"],
               'max_depth': [int(x) for x in np.linspace(start=5,stop=60,num=8)],
               'min_samples_split': [2,6,8,10,20,50,100],
               'min_samples_leaf': [1,2,5,10,20]}

print(param_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = param_grid,scoring="neg_root_mean_squared_error",cv = 5, 
                               verbose=2,
                               random_state=32, n_jobs = -1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
y_pred=rf_random.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("r_square:",metrics.r2_score(y_test,y_pred))
print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
